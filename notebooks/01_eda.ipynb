```python
import pandas as pd
from pathlib import Path
```

```python

DATA_DIR = Path("../data")  # или твоя папка с .dat

ratings = pd.read_csv(
    DATA_DIR / "ratings.dat",
    sep="::",
    engine="python",
    encoding="iso-8859-1",
    names=["user_id", "movie_id", "rating", "timestamp"],
)
users = pd.read_csv(
    DATA_DIR / "users.dat",
    sep="::",
    engine="python",
    encoding="iso-8859-1",
    names=["user_id", "gender", "age", "occupation", "zip"],
)
movies = pd.read_csv(
    DATA_DIR / "movies.dat",
    sep="::",
    engine="python",
    encoding="iso-8859-1",
    names=["movie_id", "title", "genres"],
)

print(ratings.head())
print(users.head())
print(movies.head())
```
